# Create graphs for Leakage current

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

## Import data

```python
root_path = '../../AIMSPICE/simulation result final/' 
main_FF_20 = pd.read_csv(root_path + 'main_FF_20', sep='\t', names=['t', 'v_in', 'v_sel', 'v_rw', 'v_latch', 'v_cell'], skiprows=1)
```

## Converter Functions

```python
def to_micro(x, pos):
    return f"{x * 1e6:.000f}"

def to_nano(x, pos):
    return f"{x * 1e9:.000f}"

def to_pico(x, pos):
    return f"{x * 1e12:.000f}"
```

## Plot the graph

### v for FF 20

```python
# Plot the files
plt.figure(figsize=(10, 4))
plt.title('FF at 20', fontsize=14)

plt.subplot(3, 1, 1)
plt.plot(main_FF_20['t'], main_FF_20['v_in'], label='v_in', linewidth=2)
plt.ylabel('V_in', fontsize=12)
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.xlim(0, 8e-8)

plt.subplot(3, 1, 2)
plt.plot(main_FF_20['t'], main_FF_20['v_sel'], label='v_sel', linewidth=2)
plt.ylabel('V_sel', fontsize=12)
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.xlim(0, 8e-8)

plt.subplot(3, 1, 3)
plt.plot(main_FF_20['t'], main_FF_20['v_rw'], label='v_rw', linewidth=2)
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('V_rw', fontsize=12)
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.xlim(0, 8e-8)
# Customize the plot


#plt.ylabel('Voltage ()', fontsize=12)
#plt.grid(True, which='both', linewidth=0.5)
#plt.legend(fontsize=12)
plt.tight_layout()


# Convert axis

#plt.gca().yaxis.set_major_formatter(FuncFormatter(to_nano))

# Display the plot
plt.show()
```
