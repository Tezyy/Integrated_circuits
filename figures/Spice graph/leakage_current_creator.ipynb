# Create graphs for Leakage current

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

## Import data

```python
root_path = '../../AIMSPICE/simulation result final/' 
lc_FF_000 = pd.read_csv(root_path + 'leakage_current__FF_000', sep='\t', names=['t', 'i'], skiprows=1)
lc_FF_001 = pd.read_csv(root_path + 'leakage_current__FF_001', sep='\t', names=['t', 'i'], skiprows=1)
lc_FF_100 = pd.read_csv(root_path + 'leakage_current__FF_100', sep='\t', names=['t', 'i'], skiprows=1)
lc_FF_101 = pd.read_csv(root_path + 'leakage_current__FF_101', sep='\t', names=['t', 'i'], skiprows=1)
lc_SS_000 = pd.read_csv(root_path + 'leakage_current__SS_000', sep='\t', names=['t', 'i'], skiprows=1)
lc_SS_001 = pd.read_csv(root_path + 'leakage_current__SS_001', sep='\t', names=['t', 'i'], skiprows=1)
lc_SS_100 = pd.read_csv(root_path + 'leakage_current__SS_100', sep='\t', names=['t', 'i'], skiprows=1)
lc_SS_101 = pd.read_csv(root_path + 'leakage_current__SS_101', sep='\t', names=['t', 'i'], skiprows=1)
lc_TT_000 = pd.read_csv(root_path + 'leakage_current__TT_000', sep='\t', names=['t', 'i'], skiprows=1)
lc_TT_001 = pd.read_csv(root_path + 'leakage_current__TT_001', sep='\t', names=['t', 'i'], skiprows=1)
lc_TT_100 = pd.read_csv(root_path + 'leakage_current__TT_100', sep='\t', names=['t', 'i'], skiprows=1)
lc_TT_101 = pd.read_csv(root_path + 'leakage_current__TT_101', sep='\t', names=['t', 'i'], skiprows=1)
```

## Converter Functions

```python
def to_micro(x, pos):
    return f"{x * 1e6:.000f}"

def to_nano(x, pos):
    return f"{x * 1e9:.000f}"

def to_pico(x, pos):
    return f"{x * 1e12:.000f}"
```

## Plot the graph

### Leakage current 000

```python
# Plot the files
plt.figure(figsize=(10, 6))
plt.plot(lc_FF_000['t'], lc_FF_000['i'], label='Corner FF', linewidth=2)
plt.plot(lc_SS_000['t'], lc_SS_000['i'], label='Corner SS', linewidth=2)
plt.plot(lc_TT_000['t'], lc_TT_000['i'], label='Corner TT', linewidth=2)

# Customize the plot
plt.title('Leakage Current for in=0, sel=0, rw=0', fontsize=14)
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('Current (nA)', fontsize=12)
#plt.grid(True, which='both', linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()
plt.xlim(0, 8e-8)

# Convert axis
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_nano))

# Display the plot
plt.show()
```
