## Converter Functions

```python
def to_micro(x, pos):
    return f"{x * 1e6:.000f}"

def to_nano(x, pos):
    return f"{x * 1e9:.000f}"

def to_pico(x, pos):
    return f"{x * 1e12:.000f}"
```

## Plot the graph

### v for stimuli

```python
# Plot the files
plt.figure(figsize=(10, 4))
plt.title('Stimuli Voltages', fontsize=14)

plt.xticks([])
plt.yticks([])

plt.subplot(3, 1, 1)
plt.plot(Vin_Vsel_Vop['t'], Vin_Vsel_Vop['v_in'], label='v_in', linewidth=2)
plt.ylabel('$V_{in}$', fontsize=12)
plt.xticks([])
plt.yticks([0, 1])
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.xlim(0, 8e-8)

plt.subplot(3, 1, 2)
plt.plot(Vin_Vsel_Vop['t'], Vin_Vsel_Vop['v_sel'], label='v_sel', linewidth=2)
plt.ylabel('$V_{sel}$', fontsize=12)
plt.xticks([])
plt.yticks([0, 1])
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.xlim(0, 8e-8)

plt.subplot(3, 1, 3)
plt.plot(Vin_Vsel_Vop['t'], Vin_Vsel_Vop['v_rw'], label='v_rw', linewidth=2)
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('$V_{rw}$', fontsize=12)
plt.xticks(np.arange(0, 9e-8, 5e-9))
plt.yticks([0, 1])
plt.gca().xaxis.set_major_formatter(FuncFormatter(to_nano))
plt.xlim(0, 8e-8)

# Customize the plot
plt.tight_layout()

# Display the plot
plt.show()
```
