# Create graphs for Aimspice Results

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

## Import data

```python
root_path = '../../AIMSPICE/simulation result final/' 
lc_FF_000 = pd.read_csv(root_path + 'leakage_current__FF_000', sep='\t', names=['Time (s)', 'Current_1 (A)'], skiprows=1)
lc_FF_001 = pd.read_csv(root_path + 'leakage_current__FF_001', sep='\t', names=['Time (s)', 'Current_2 (A)'], skiprows=1)
lc_FF_100 = pd.read_csv(root_path + 'leakage_current__FF_100', sep='\t', names=['Time (s)', 'Current_3 (A)'], skiprows=1)
lc_FF_101 = pd.read_csv(root_path + 'leakage_current__FF_101', sep='\t', names=['Time (s)', 'Current_4 (A)'], skiprows=1)
```

## Plot the graph

```python
# Plot the data from both files
plt.figure(figsize=(10, 6))
plt.plot(lc_FF_000['Time (s)'], lc_FF_000['Current_1 (A)'], label='Leakage Current FF_000', linewidth=2)
plt.plot(lc_FF_001['Time (s)'], lc_FF_001['Current_2 (A)'], label='Leakage Current FF_001', linewidth=2)
plt.plot(lc_FF_100['Time (s)'], lc_FF_100['Current_3 (A)'], label='Leakage Current FF_100', linewidth=2)
plt.plot(lc_FF_101['Time (s)'], lc_FF_101['Current_4 (A)'], label='Leakage Current FF_101', linewidth=2)

# Customize the plot
plt.title('Leakage Current for FF', fontsize=14)
plt.xlabel('Time (ps)', fontsize=12)
plt.ylabel('Current (nA)', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()
plt.xlim(0, 0.075e-8)

def to_nanoseconds(x, pos):
    return f"{x * 1e9:.0f} ns"

def to_picoseconds(x, pos):
    return f"{x * 1e12:.0f}"

def to_nanoampere(x, pos):
    return f"{x * 1e9:.0000f}"

plt.gca().xaxis.set_major_formatter(FuncFormatter(to_picoseconds))
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_nanoampere))

# Display the plot
plt.show()
```
